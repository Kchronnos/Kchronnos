```python
#search full archive permite buscar tweets más antiguos de 30 días

results = tweepy.Cursor(api.search_full_archive, 'tfm', '#bitcoin', fromDate='202104250000',
                            toDate='202104260000').items()


#results = api.search_full_archive('tfm', '#bitcoin', fromDate='202104250000', toDate='202104260000', maxResults=100)
```

```python

# create list to append tweets to
tweets = []

# append all tweet data to list
for tweet in results:
    tweets.append(tweet)

# convert 'tweets' list to pandas.DataFrame
tweets_df = pd.DataFrame(vars(tweets[i]) for i in range(len(tweets)))

# define file path (string) to save csv file to
FILE_PATH = '25-04-2021.csv'

# use pandas to save dataframe to csv
tweets_df.to_csv(FILE_PATH)
```
